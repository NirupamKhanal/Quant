# Geometric Brownian Motion
$S_{t} = S_{0}e^{(\mu-\frac{{\sigma}^2}{2})t+\sigma W_{t}}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0.1 # drift coefficient 
n = 100 # no. of steps
T = 1 # Time in years
M = 100 # number of simulations
S0 = 100 # initial stock price
sigma = 0.3 # volatility 
```

## Simulating GMB Paths

```python
dt = T/n
St = np.exp((mu - sigma**2/2)*dt + sigma*np.random.normal(0, np.sqrt(dt), size=(M,n)).T)
# initialization 
St = np.vstack([np.ones(M), St])
St = S0*St.cumprod(axis=0)
St[1]
```

```python
# Plotting
time = np.linspace(0, T, n+1)
tt = np.full(shape=(M, n+1), fill_value=time).T

plt.plot(tt, St)
plt.xlabel("Years $(t)$")
plt.ylabel("Stock Price $(S_{t})$")
plt.title("Realizations of Geometric Brownian Motion\n")
plt.show()
```
